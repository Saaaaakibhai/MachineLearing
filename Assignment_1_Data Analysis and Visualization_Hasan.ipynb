### For Save Image

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "ImagesForAssignment/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

## Question No 01 
=> Region wise life expectancy

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#creating a count plot using sns

plt.figure(figsize=(18,10))
sns.barplot(data=df, x=df.Region , y=df.Life_expectancy)
plt.title ('Region wish life expectancy')
plt.xlabel("Region")
plt.ylabel("Life_expectancy")
save_fig("Question no 01")
plt.show()
```

## Question no 02
=> Alcohol_consumption by year

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Creating a swarm plot that shows the relation between a categorial and continuous variable
plt.figure(figsize=(15,10))
sns.swarmplot(data=df,x='Year',y='Alcohol_consumption',palette='Set2',size=3)
plt.title('Swam Plot Alcohol_consumption vs BMI')
plt.xlabel('Year')
plt.ylabel('Alcohol_consumption')
#show the fig
save_fig("Swarm Plot Alcohol_consumption vs BMI")
#show the plot
plt.show()
```

In ths swarmplot we can see that every year this Alcohol_consumption rate being same every year
