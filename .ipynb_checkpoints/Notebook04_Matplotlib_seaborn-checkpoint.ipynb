# Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd


#creating a sample dataframe
data={
    'Categories': ['Category A','Category B','Category C','Category D'],
    'Values' : [26,49,30,56]
}
df=pd.DataFrame(data)

#plotting the bar chart using matplotlib bar function
plt.bar(df['Categories'],df['Values'])
#Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category va Values')
#add grid
plt.grid()
#display the chart
plt.show()
```

## Bar Chart


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


#creating a sample dataframe
data={
    'Categories': ['Category A','Category B','Category C','Category D'],
    'Values' : [26,49,30,56]
}
df=pd.DataFrame(data)

#plotting the bar chart using matplotlib bar function
plt.bar(df['Categories'],df['Values'])

#Ploting the bar plot using sns
sns.barplot(x=df.Categories,y=df.Values)

#Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category va Values')
#add grid
#plt.grid()
#display the chart
plt.show()
```

# Choosing Seaborn Color Palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Sample df 
data={
    'Categories':['A','B','C','D'],
    'Values':[26,49,30,56]
}
df=pd.DataFrame(data)
#list of color palettes
palette_list=['deep','muted','bright','pastel','dark','colorblind']
#set figure size
plt.figure(figsize=(12,8))

for i,palette in enumerate(palette_list):
    plt.subplot(2,3,i+1)
    sns.barplot(data=df,x=df.Categories,y=df.Values,palette=palette)
    plt.title(f'Palette: {palette}')

#Applying tight layout
plt.tight_layout()
#show the plot
plt.show()
```

# Pie Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Sample df 
data={
    'Categories':['A','B','C','D'],
    'Values':[26,49,30,56]
}
df=pd.DataFrame(data)

#set fig size
plt.figure(figsize=(6,6))
#creating a pie chart
plt.pie(df['Values'],labels=df['Categories'],
        autopct='%1.1f%%',
        startangle=90,
       colors=['salmon','orange','khaki','red'])
plt.title('Pie Chart')
#save the figure
save_fig('Matplotlib pie chart')
#show plot 
plt.show()
```

# Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Creating a custom DataFrame
data={
    'Stroke':['Yes','No','Not Sure','Not Sure',"Sure","Maybe","Yes",'No','Yes','No']
}
df=pd.DataFrame(data)

#Creating a count plot using sns
plt.figure(figsize=(8,6))
sns.countplot(data=df,x=df.Stroke)
plt.title('Stroke Count')
plt.xlabel("Stroke Status")
plt.ylabel("Count")


#use grid
plt.grid()
#save the figure
save_fig('Stroke value count plot')
#show the fig
plt.show()
```
