# Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd


#creating a sample dataframe
data={
    'Categories': ['Category A','Category B','Category C','Category D'],
    'Values' : [26,49,30,56]
}
df=pd.DataFrame(data)

#plotting the bar chart using matplotlib bar function
plt.bar(df['Categories'],df['Values'])
#Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category vs Values')
#add grid
plt.grid()
#display the chart
plt.show()
```

## Bar Chart


```python
#Using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#creating a sample dataframe
data={
    'Categories': ['Category A','Category B','Category C','Category D'],
    'Values' : [26,49,30,56]
}
df=pd.DataFrame(data)

#plotting the bar chart using matplotlib bar function
plt.bar(df['Categories'],df['Values'])

#Ploting the bar plot using sns
sns.barplot(x=df.Categories,y=df.Values)

#Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category va Values')
#add grid
plt.grid()
#display the chart
plt.show()
```

# Choosing Seaborn Color Palettes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Sample df 
data={
    'Categories':['A','B','C','D'],
    'Values':[26,49,30,56]
}

df=pd.DataFrame(data)

#list of color palettes
palette_list=['deep','muted','bright','pastel','dark','colorblind']
#set figure size
plt.figure(figsize=(12,8))

for i,palette in enumerate(palette_list):
    plt.subplot(2,3,i+1)
    sns.barplot(data=df,x=df.Categories,y=df.Values,palette=palette)
    plt.title(f'Palette: {palette}')

#Applying tight layout
plt.tight_layout()
#show the plot
plt.show()
```

# Function to save figures

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "Images/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

# Pie Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Sample df 
data={
    'Categories':['A','B','C','D'],
    'Values':[26,40,30,35]
}
df=pd.DataFrame(data)
#set fig size
plt.figure(figsize=(6,6))
#creating a pie chart
plt.pie(df['Values'],
        labels=df['Categories'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['salmon','orange','khaki','lightcyan'])

plt.title('Pie Chart')
#save the figure
save_fig('Matplotlib pie chart')
#show plot 
plt.show()
```

# Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#Creating a custom DataFrame
data={
    'Stroke':['Yes','No','Not Sure','Not Sure',"Sure","Maybe","Maybe","Yes",'Yes','No']
}
df=pd.DataFrame(data)

#Creating a count plot using sns
plt.figure(figsize=(8,6))
sns.countplot(data=df,x=df.Stroke)
plt.title('Stroke Count')
plt.xlabel("Stroke Status")
plt.ylabel("Count")
#use grid
plt.grid()
#save the figure
save_fig('Stroke value count plot')
#show the fig
plt.show()
```

# Hist Plot

## Difference Between Count and Hist Plot
The main differences between a count plot and a histplot lie in the types of data they are used to visualize.

Key Differences: Data Type: Count plots deal with categorical data and show the count of occurrences within each category. Histplots handle numeric data, showing the distribution and frequency of values.

Variable Representation: Count plots display categorical variables as bars with the count on the y-axis. Histplots show the distribution of numeric values as bars representing bins on the x-axis.

Usage: Count plots are suitable for categorical variables' frequency comparison, while histplots are used to understand the distribution of continuous variables.

## Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#Creating a dataset
df=pd.DataFrame({
    'Date':pd.date_range('2023-01-01',periods=90),
    'Value':np.random.uniform(low=0,high=100,size=90)
})
#Creating a line plot
plt.figure(figsize=(10,8))
sns.lineplot(data=df,x='Date',y='Value',color='blue')
plt.title('Time plot of Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
#Rotating xlabel
plt.xticks(rotation=45)
#show the fig
save_fig("lineplot01")
#show the plot
plt.show()
```

## Swarn Plot 

```python
import pandas as pd
#import dataset
df=pd.read_csv('Datasets/healthcare-dataset-stroke-data.csv')
df.info() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Creating a swarm plot that shows the relation between a categorial and continuous variable
plt.figure(figsize=(8,6))
sns.swarmplot(data=df,x='stroke',y='bmi',palette='Set2',size=3)
plt.title('Swam Plot Stroke vs BMI')
plt.xlabel('Stroke Status')
plt.ylabel('BMI')
#show the fig
save_fig("Swarm Plot Stroke vs BMI")
#show the plot
plt.show()
```

```python
#Boxplot
sns.boxplot(x=df.bmi)
plt.xlabel('BMI')
plt.title('Box Plot')
plt.grid()
plt.show()
```
